```python
import hmac
import hashlib
import json

def generate_leadlens_signature(payload: dict, secret_key: str) -> str:
    # 1. Serialize the payload to a compact JSON string (no whitespace outside strings)
    # This matches exactly how the FastAPI server stringifies payloads
    payload_str = json.dumps(payload, separators=(',', ':'))
    
    # 2. Compute HMAC-SHA256 signature
    signature = hmac.new(
        secret_key.encode('utf-8'),
        payload_str.encode('utf-8'),
        hashlib.sha256
    ).hexdigest()
    
    return signature

# ==========================================
# Example Usage:
# ==========================================
if __name__ == "__main__":
    # Prompt user for inputs
    secret = input("Enter your secret key (e.g. shoption-dev-123): ").strip()
    
    # Example payload matching the one synced
    sample_payload = {
        "event": "call.synced",
        "timestamp": "2026-06-19T05:30:00.123456Z",
        "org_id": "738519f3-76a9-4ec8-a302-82709985a20e",
        "data": {
            "phone_number": "+9190491890",
            "call_type": "Inming",
            "call_status": "Miss Call",
            "duration_seconds": 0,
            "timestamp": "19-Jun-2026 11:00",
            "system_call_id": "11923",
            "employee_id": "EMP-001",
            "system_id": "862473"
        }
    }
    
    sig = generate_leadlens_signature(sample_payload, secret)
    print("\nGenerated Signature:")
    print(sig)
```
